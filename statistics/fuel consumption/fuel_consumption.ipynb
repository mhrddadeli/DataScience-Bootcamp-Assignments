خروجی شما ماکسیمم درصدی است که می‌توانید با آن ادعای شرکت فولکس واگن را رد کنید. دقت کنید که تفاوت پاسخ شما با جواب واقعی باید کمتر از

```python
import numpy as np
import scipy as sp
import math
```

```python
arr = np.array([[85, 86, 88, 75, 78, 94, 98, 79, 71, 80],
       [91, 92, 93, 85, 87, 84, 82, 88, 95, 96],
       [79, 78, 88, 94, 92, 85, 83, 85, 82, 81]])
```

```python
country_means = np.array([np.mean(country) for country in arr])
```

```python
def solve(arr):
    arr = np.array(arr)

    k = len(arr)
    total_n = len(arr.flatten())
    n_countries = np.array([len(country) for country in arr])
    grand_mean = np.mean(arr)
    country_means = np.array([np.mean(country) for country in arr])

    ssb = np.sum((n_countries * np.power(country_means - grand_mean, 2)))  # Sum of Squares Between 
    ssw = np.sum(np.power(arr - country_means.reshape(k, 1), 2)) # Sum of Squares Within

    # Degrees of Freedom
    df_ssb = k - 1
    df_ssw = total_n - k

    msb = ssb / df_ssb # Between variance
    msw = np.sum(ssw) / df_ssw # Within variance
    f = msb / msw

    p_value = 1 - sp.stats.f.cdf(f, df_ssb, df_ssw)
    confidence = (1 - p_value) * 100
    
    return confidence 
```
